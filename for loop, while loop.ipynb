```python
s="ineuron"
vowel=[]
no_vowel=[]
v="AaEeIiOoUu"
for i in s:
    if i not in v:
        no_vowel.append(i)
    else:
        vowel.append(i)
print("vowels:",vowel)
print("consonants:",no_vowel)
```

```python
dic={"ineuron":
               {
                   "a":14,
                   "b":10,
                   "c":4
               },
     "course": {
                   "d":45,
                   "e":34,
                   "f":1
               }
    }
```

```python
l=[]
for i in dic.values():
    for j in i.values():
        l.append(j)
    print(max(l))
```

```python
g=[]
d={"ineuron":{"a":14, "b":10,"c":4},"course": {"d":45,"e":34,"f":1},"g":34,"h":[46,6,7,8,9,3],"i":(45,34,2),"k":"sudh"}
for i in d.values():
    if type(i)==dict:
        g.append(max(i.values()))
    elif type(i)==list:
        g.append(max(i))
    elif type(i)==tuple:
        g.append(max(i))
    elif type(i)==str:
        pass
    else:
        g.append(i)
print(max(g))       
```

```python
import random

r = random.randint(0,100)

attempts = 0
while True:
    attempts += 1
    Guess=int(input("Guess the number : "))
    if Guess>r:
          print("Number too large")
      
    elif Guess<r:
          print("Number is too small")
    elif Guess == 0:
          break
    else:print("Congratulations! You guessed it right")
print("attempt:",attempts)
            
```
